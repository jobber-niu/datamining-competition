```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import time
```

```python
def B():
    path = './不同地区数据'
    file_names = os.listdir(path)

    # 将所有数据表读入到一个字典中，键为 data1 到 data5
    data_dict = {}
    for i, file_name in zip(range(1, 6), file_names):
        df = pd.read_csv(os.path.join(path, file_name))
        data_dict["data{}".format(i)] = df
        time.sleep(2)
    return  data_dict
data_dict = B()
```

```python
df = data_dict['data5']
plt.subplot(121)
sns.scatterplot(x='item_price',y='ord_qty',data=df)
time.sleep(5)
plt.subplot(122)
ax = sns.scatterplot(x='item_price',y='ord_qty',data=df)
plt.show(ax)
time.sleep(5)
```

```python
fig = plt.figure(figsize = (50,50))
# plt.rcParams.updata({'font.size':18})
sns.scatterplot(x='item_price',y='ord_qty',data=df)
plt.savefig('./niu_bi.png')
```

```python
i = 1
fig = plt.figure(figsize = (40,40))
for i in data_dict:
    # print(i)
    df = data_dict[i]
    sns.scatterplot(x='item_price',y='ord_qty',data=df)
#     time.sleep(5)
    plt.savefig('./'+i+'.png')
    plt.close()
# 数据杂乱的原因可能在于没有将大类分开，继续处理
```
